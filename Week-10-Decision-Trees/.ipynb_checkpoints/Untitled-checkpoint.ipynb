```python
import pycountry
import numpy as np
```

```python
countries = pycountry.countries
countries = [country.name for country in countries]
countriesFromText = ['Jamaica', 'United States', 'United Kingdom']
```

```python
countryList = [0] * len(countries)
for i, country in enumerate(countries):
    print(i, country)
    countryList[i] = countriesFromText.count(country)
    
countryList = np.array(countryList)
countryList
```
