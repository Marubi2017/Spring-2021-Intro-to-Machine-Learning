```python
import glob # for reading files
import pandas as pd
```

```python
datafiles = glob.glob('data/PRSA/*')

data = []

for file in datafiles:
    data.append(pd.read_csv(file))
    
data = pd.concat(data)

data
```

```python
# this line gets rid of the 'No' column
data = data.drop(columns = ['No'])

# this drops unavailable(n/a) data
data = data.dropna()

data
```

```python
# gonna represent the station column as categorical variables
data = pd.get_dummies(data, columns = ['station'])

data
```
